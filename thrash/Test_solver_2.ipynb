```python
%matplotlib

from odeintw import odeintw
import numpy as np
import matplotlib.pyplot as plt

def time_dep(t):
    return (np.sin(t))

coefficients = {'coeff1' : lambda x: time_dep(x)}

def zfunc(z, t, L, M):
    z1, z2 = z
    w = wc + wm*Am*coefficients['coeff1'](t)
    print(t,w)
    return [(-1j * w) * z1 + L * z2, M*z2]

def zjac(z, t, L, M):
    z1, z2 = z
    w = wc + wm*Am*coefficients['coeff1'](t)
    jac = np.array([[(-1j * w), L], [0, M]])
    return jac

print(coefficients['coeff1'](1))
```

```python
# Initial conditions.
z0 = np.array([0+0j, 1+0j])

# Auxiliar parameters

g = 2 * np.pi * 1
wc = 2 * np.pi* 1000
wm = 2 * np.pi* 0.1
wx = 2 * np.pi* 1000
Am = 2 * np.pi* 250


# Parameters.
L = -1j * g
M = -1j * wx


# Desired time samples for the solution.
t0 = np.linspace(0, 50, 101)

z, infodict = odeintw(zfunc, z0, t0, args=(L, M), Dfun=zjac, full_output=True)
```

```python
print(z[3])
```

```python
color1 = (0.5, 0.4, 0.3)
color2 = (0.2, 0.2, 1.0)
plt.plot(t0, z[:, 0].real, color=color1, label='z1.real', linewidth=1.5)
plt.plot(t0, z[:, 0].imag, '--', color=color1, label='z1.imag', linewidth=2)
plt.plot(t0, z[:, 1].real, color=color2, label='z2.real', linewidth=1.5)
plt.plot(t0, z[:, 1].imag, '--', color=color2, label='z2.imag', linewidth=2)
plt.xlabel('t')
plt.grid(True)
plt.legend(loc='best')
plt.show()
```
